# 1.Maximum value from given matrix

```python
# create 2x2 matrix and find the max
import numpy as np
m=np.matrix('[23,12;34,56]')
mx=m.max()
print(mx)
```

# 2.Minimum value from given matrix

```python
# finding min()
m=np.matrix('[23,12;34,56]')
mx=m.min()
print(mx)
```

# 3.Find number of rows and columns of a given matrix

```python
# finding no of rows and column
#using shape()
n=np.matrix([[23,45,61],[1,23,41]])
rows,columns=n.shape
print("Rows:",rows)
print("Column:",columns)
```

```python
#using indexing[]
n=np.matrix([[23,45,61],[1,23,41]])
rows=n.shape[0]  # 0 gives number of rows
columns=n.shape[1] # 1 gives number of columns
print("Rows:",rows)
print("Column:",columns)
```

```python
#using reshape()
import numpy as np
m=np.arange(0,12).reshape((4,3))
print(m)
rows,columns=m.shape
print("Rows:",rows)
print("Column:",columns)
```

# 7.Adding and Subtracting Matrices

```python
a=np.matrix('[1,2,3]')
b=np.matrix('[4,5,6]')

print("Matrix a is :\n",a)
print("Matrix b is :\n",b)

c=np.add(a,b)
print("Result :\n",c)
```

```python
import numpy as np
a=np.matrix('[10,20,30]')
b=np.matrix('[4,5,6]')

print("Matrix a is :\n",a)
print("Matrix b is :\n",b)

d=np.subtract(a,b)
print("Result :\n",d)
```

# 8.Ways to add row/columns

```python
#np.append
a=np.matrix([[1,2,3],[4,5,6],[7,8,9]])
b=np.array([[11],[12],[13]])  
c=np.append(a,b,axis=1)
print(c)
```

```python
#np.insert
a=np.matrix([[1,2,3],[4,5,6],[7,8,9]])
b=np.array([11,12,13])
n = a.shape[0]  
c=np.insert(a,n,b,axis=0)
print(c)
```

# 9.Matrix Multiplication

```python
import numpy as np
a = [[1, 2], [2, 3]]
b = [[4, 5], [6, 7]]

print("Matrix a is :\n",a)
print("Matrix b is :\n",b)

c=np.dot(a,b)
print("Result :\n",c)
```

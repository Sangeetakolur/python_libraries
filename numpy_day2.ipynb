# indexing and slicing

```python
arr=np.array([50,60,70,80,90,55])
```

```python
print(arr[0])
```

```python
print(arr[-1])
```

```python
arr[0:5]
```

```python
arr[0:6]
```

```python
arr[:4] #equal to arr[0:4]
```

```python
arr[2:] #equal to [2:6]
```

```python
my_list=[1,2,3,4,5,6]
print(my_list)
new_list=my_list[0:4]
print(new_list)
new_list[0]=100
print(new_list)
print(my_list)
```

```python
my_array=np.array([1,2,3,4,5,6])
print(my_array)
new_array=my_array[0:4]
print(new_array)
new_array[0]=100
print(new_array)
print(my_array)
```

# creation of numpy array

```python
import numpy as np

a = np.array([1,2,3,4])
print(a)
```

```python
a1=np.array([1,2,34,4])
print(a1)
print(*a1)
print(a1.ndim)
```

```python
a2=np.array([[1,2],[3,4]])
print(a2)
```

```python
a3=np.array([["a","b","c"],[1,2,3],[4,5,6]])
print(a3)
print(*a3)
print(a3.ndim)
```

```python
a4=np.array([34])
print(a4)
print(*a4)
print(a4.ndim)
```

# declaration of array (data types in numpy)

```python
a5=np.array([1,2,3],dtype='i4')
a5
```

```python
a5=np.array([1,2,3],dtype='int32')
a5
```

```python
a5=np.array([1,2,3],dtype='S')
a5
```

```python
a5=np.array([1,'2',3.5])
a5
```

```python
a5=np.array([1,'2','3.5'],dtype='<U32')
a5
```

```python
import numpy as np

a = np.array([True, False, True], dtype='bool')
print(a)
print(a.dtype)
```

```python
b=np.array(['apple','banana','pineapple'])
print(b)
print(b.dtype)
```

```python
# Changing Data Type
c=np.array([1.2,2.2,3.2])
print(c)
d=c.astype(int)
print(d)
print(d.dtype)
```
